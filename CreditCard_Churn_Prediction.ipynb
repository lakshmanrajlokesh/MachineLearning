```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import gradio as gr

# Load and clean data
churndata = pd.read_csv("Credit card churn.csv")
churndata = churndata.iloc[:, :-2]
churndata.replace(0, np.nan, inplace=True)
churn_data = churndata.fillna(churndata.mean(numeric_only=True))

# Encode categorical columns
categorical_columns = ["Attrition_Flag", "Gender", "Education_Level", "Marital_Status",
                       "Income_Category", "Card_Category"]
label_encoders = {}
for column in categorical_columns:
    le = LabelEncoder()
    churn_data[column] = le.fit_transform(churn_data[column])
    label_encoders[column] = le

# Drop CLIENTNUM
churn_data.drop(columns=["CLIENTNUM"], inplace=True)

# Define features and target
X = churn_data.drop(columns=["Attrition_Flag"])
y = churn_data["Attrition_Flag"]

# Standardize numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

# Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.10, random_state=30)

# Train Random Forest model
rf_clf = RandomForestClassifier(n_estimators=500, max_depth=15, min_samples_split=5, min_samples_leaf=2, random_state=42)
rf_clf.fit(X_train, y_train)

# Capture feature names
feature_names = X.columns.tolist()

# Gradio interface
def predict_churn(*inputs):
    # Prepare input DataFrame
    input_dict = dict(zip(feature_names, inputs))
    input_df = pd.DataFrame([input_dict])

    # Encode categorical features
    for col in categorical_columns:
        if col != "Attrition_Flag":
            le = label_encoders[col]
            input_df[col] = le.transform([input_df[col][0]])

    # Scale numeric features
    input_scaled = scaler.transform(input_df)
    input_scaled_df = pd.DataFrame(input_scaled, columns=feature_names)

    # Predict
    pred = rf_clf.predict(input_scaled_df)[0]
    return "Churn" if pred == 1 else "Not Churn"

# Define inputs
input_components = [
    gr.Radio(["Male", "Female"], label="Gender"),
    gr.Dropdown(label_encoders["Education_Level"].classes_.tolist(), label="Education Level"),
    gr.Dropdown(label_encoders["Marital_Status"].classes_.tolist(), label="Marital Status"),
    gr.Dropdown(label_encoders["Income_Category"].classes_.tolist(), label="Income Category"),
    gr.Dropdown(label_encoders["Card_Category"].classes_.tolist(), label="Card Category"),
    gr.Number(label="Customer_Age"),
    gr.Number(label="Dependent_count"),
    gr.Number(label="Months_on_book"),
    gr.Number(label="Total_Relationship_Count"),
    gr.Number(label="Months_Inactive_12_mon"),
    gr.Number(label="Contacts_Count_12_mon"),
    gr.Number(label="Credit_Limit"),
    gr.Number(label="Total_Revolving_Bal"),
    gr.Number(label="Avg_Open_To_Buy"),
    gr.Number(label="Total_Amt_Chng_Q4_Q1"),
    gr.Number(label="Total_Trans_Amt"),
    gr.Number(label="Total_Trans_Ct"),
    gr.Number(label="Total_Ct_Chng_Q4_Q1"),
    gr.Number(label="Avg_Utilization_Ratio")
]

# Map inputs to order expected by model
def reorder_inputs(*args):
    gender, edu, marital, income, card, *nums = args
    input_dict = {
        "Gender": gender,
        "Education_Level": edu,
        "Marital_Status": marital,
        "Income_Category": income,
        "Card_Category": card,
        "Customer_Age": nums[0],
        "Dependent_count": nums[1],
        "Months_on_book": nums[2],
        "Total_Relationship_Count": nums[3],
        "Months_Inactive_12_mon": nums[4],
        "Contacts_Count_12_mon": nums[5],
        "Credit_Limit": nums[6],
        "Total_Revolving_Bal": nums[7],
        "Avg_Open_To_Buy": nums[8],
        "Total_Amt_Chng_Q4_Q1": nums[9],
        "Total_Trans_Amt": nums[10],
        "Total_Trans_Ct": nums[11],
        "Total_Ct_Chng_Q4_Q1": nums[12],
        "Avg_Utilization_Ratio": nums[13]
    }
    return predict_churn(*input_dict.values())

# Launch Gradio app
demo = gr.Interface(fn=reorder_inputs,
                    inputs=input_components,
                    outputs="text",
                    title="Credit Card Churn Predictor",
                    description="Enter customer data to predict if they are likely to churn.")

demo.launch()
```
